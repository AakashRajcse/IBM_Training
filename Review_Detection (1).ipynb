```python
# Label create
df['label'] = df['overall'].apply(lambda x: 1 if x <= 2 else 0)
```

```python
# Optional: remove neutral
df = df[df['overall'] != 3]

df.head()
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    return text

df['clean_review'] = df['reviewText'].apply(clean_text)
```

```python
#Text Cleaning (NLP)

df['clean_review'] = df['reviewText'].apply(clean_text)
```

## Stopwords + Tokenization

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    words = word_tokenize(text)
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df['processed_review'] = df['clean_review'].apply(preprocess)
```

## Text → Numbers (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['processed_review'])
y = df['label']
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Model Training (Naive Bayes 🔥)

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

## Prediction + Accuracy

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Test Your Own Review

```python
def predict_review(text):
    text = clean_text(text)
    text = preprocess(text)
    vec = vectorizer.transform([text])
    pred = model.predict(vec)
    
    return "Fake Review " if pred[0] == 1 else "Genuine Review "

print(predict_review("This product is very bad and useless"))
```

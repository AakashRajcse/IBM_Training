## Self Data

```python
data = {
    'Size_sqft': [600, 800, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400],
    'Price_1000s': [150, 180, 210, 240, 270, 300, 330, 360, 390, 420]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["Price_1000s"], df["Size_sqft"])
plt.xlabel("Price")
plt.ylabel("Size")
plt.show()
```

```python
X = df[['Price_1000s']]      
y = df['Size_sqft']  
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
X_test.iloc[0]
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
plt.scatter(df["Size_sqft"], df["Price_1000s"])
plt.plot(X_train, lr.predict(X_train), color = "black")
plt.xlabel("Size")
plt.ylabel("Price")
plt.show()
```

```python
m = lr.coef_
b = lr.intercept_

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
Price = m * 270. + b
```
